```python
from sklearn import datasets
import numpy as np
```

```python
iris = datasets.load_iris()
X = iris.data[:, 2:]
y = iris.target
print('Number of samples:', X.shape[0])
print('Class labels:', np.unique(y))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
X_train
```

```python
print('Labels counts in y:', np.bincount(y))
print('Label counts in y_train:', np.bincount(y_train))
print('Label counts in y_test:', np.bincount(y_test))
```

### np.bincount解釋

np.bincount 會統計每個非負整數在陣列中出現的次數。輸出陣列的索引代表「整數值」,對應的元素代表「該整數出現幾次」。

輸入是 [0, 1, 1, 3, 2, 1, 7],最大值是 7,所以輸出長度是 7+1 = 8(索引 0 到 7):

| 索引(整數值) | 出現次數 | 說明 |
|---|---|---|
| 0 | 1 | `0` 出現 1 次 |
| 1 | 3 | `1` 出現 3 次 |
| 2 | 1 | `2` 出現 1 次 |
| 3 | 1 | `3` 出現 1 次 |
| 4 | 0 | 沒出現 |
| 5 | 0 | 沒出現 |
| 6 | 0 | 沒出現 |
| 7 | 1 | `7` 出現 1 次 |

所以結果就是 array([1, 3, 1, 1, 0, 0, 0, 1])。

在 scikit-learn 的情境(例如書中第 3 章)常用它來計算各類別標籤的樣本數,例如 np.bincount(y) 可以快速看出每個 class 有多少筆資料;隨機森林、bagging 的多數決投票也常搭配 np.argmax(np.bincount(...)) 找出得票最多的類別。

```python
np.bincount(np.array([0, 1, 1, 3, 2, 1, 7]))
```

### 特徵縮放

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
X_train_std
```

### 感知器模型

大多數預設使用「一對餘」(one-vs.-rest, OvR)方法，來支援「多類別分類」

```python
from sklearn.linear_model import Perceptron

ppn = Perceptron(eta0=0.1, random_state=1)
ppn.fit(X_train_std, y_train)
y_pred = ppn.predict(X_test_std)

from sklearn.metrics import accuracy_score
print('Misclassified samples: %d' % (y_test != y_pred).sum())
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
print('Accuracy: %.2f' % ppn.score(X_test_std, y_test)) # score會呼叫predict和accuracy_score
```
